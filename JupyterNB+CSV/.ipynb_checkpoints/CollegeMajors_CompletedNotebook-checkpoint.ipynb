```python
#import libraries needed
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_AllAges = pd.read_csv ('all-ages.csv') #import csv file
df_AllAges.head() #view first five lines
```

```python
#get general info on dataframe (how many columns, names of columns, 
#how many values in each column, data value type)
df_AllAges.info()
```

```python
#check if there are any null values
pd.isnull(df_AllAges).sum()
```

```python
#testing some functions, learning about data
df_AllAges["Unemployed"].min()
```

```python
df_AllAges["Employed"].corr(df_AllAges["Total"]) #correlation between two specific columns
```

```python
df_AllAges.corr() #chart of correlation between all columns
```

```python
df_AllAges.groupby(['Major_category'])[['Median']].mean()
```

```python
#Graphing data as a bar graph

#creating variables for the columns I want to look at, saved as a series
allagesmajor = df_AllAges["Major_category"] 
allagesunemployment = df_AllAges["Unemployment_rate"] 

#converting two series into dictionary
major_unemployment = dict(zip(allagesmajor,allagesunemployment)) 

#plotting dictionary: x value is the length of dict, y value is values of dict (unemployment rate)
plt.bar(range(len(major_unemployment)), list(major_unemployment.values()), align='center') 

#adding labels to x axis, using keys of dict (major category names)
plt.xticks(range(len(major_unemployment)), list(major_unemployment.keys()), rotation=90) 

plt.show() #needed in matplotlib to show the plot
```
